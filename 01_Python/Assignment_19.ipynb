<h3 style='text-align: left;'> 1. Explain GET and POST methods. </h3>

<p> <b> Ans:- <br><br> 1) GET method: <br><br> It is the most common method which can be used to send data in the unencrypted form to the server. <br> <br> 2) POST method: <br><br> It is used to send the form data to the server. The server does not cache the data transmitted using the post method. <br><br> POST means through body I am sending you data and GET means through URL I am sending you data. POST method is more secured because data is not been seen while in GET data is been seen in URL. </b> </p>

<p> <b> Example of POST and GET Method: <br> <br> In this example we have HTML5 and CSS3 files and methods have been use to link form were we perform some basic operation of maths and we have render data through python file. </b> </p> 

<p> <b> app.py </b> </p>

```python
from flask import Flask, render_template, request, jsonify

app = Flask(__name__)


@app.route("/", methods = ['GET', 'POST'])
def home_page():
    return render_template('index.html')

@app.route('/math', methods = ['POST'])
def math_operation():
    if(request.method == 'POST'):
        ops = request.form['operation']    
        num1 = int(request.form['num1'])
        num2 = int(request.form['num2'])
        if(ops == 'add'):
            r = num1 + num2
            result = 'The addition of ' + str(num1) + ' and ' + str(num2) + ' is ' + str(r)

        if(ops == 'subtract'):
            r = num1 - num2
            result = 'The subtraction of ' + str(num1) + ' and ' + str(num2) + ' is ' + str(r)

        if(ops == 'multiply'):
            r = num1 * num2
            result = 'The multiplication of ' + str(num1) + ' and ' + str(num2) + ' is ' + str(r)

        if(ops == 'divide'):
            r = num1 / num2
            result = 'The division of ' + str(num1) + ' and ' + str(num2) + ' is ' + str(r)

        return render_template('result.html',result = result)    


if __name__ == "__main__":
    app.run(host = "0.0.0.0")    
```

```python
from flask import Flask, render_template, request, jsonify

app = Flask(__name__)

@app.route('/',methods = ['GET', 'POST'])
def home_page():
    return render_template('index.html')

@app.route('/postman_data',methods = ['POST'])
def math_operation():
    if(request.method == 'POST'):
        ops = request.json['operation']    
        num1 = int(request.json['num1'])
        num2 = int(request.json['num2'])

        if(ops == 'add'):
            r = num1 + num2
            result = 'The addition of ' + str(num1) + ' and ' + str(num2) + ' is ' + str(r)

        if(ops == 'subtract'):
            r = num1 - num2
            result = 'The subtraction of ' + str(num1) + ' and ' + str(num2) + ' is ' + str(r)
        
        if(ops == 'multiply'):
            r = num1 * num2
            result = 'The multiplication of ' + str(num1) + ' and ' + str(num2) + ' is ' + str(r)

        if(ops == 'divide'):
            r = num1 / num2
            result = 'The division of ' + str(num1) + ' and ' + str(num2) + ' is ' + str(r)
            
        return jsonify(result)
    
if __name__ == "__main__":
    app.run(host = "0.0.0.0")
```
