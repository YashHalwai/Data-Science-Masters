<h3 style='text-align: left;'> 1. What is Flask Framework? What are the advantages of Flask Framework? </h3>

<h3 style='text-align: left;'> 2.  Create a simple Flask application to display ‘Hello World!!’. Attach the screenshot of the output in Jupyter Notebook. </h3>

```python
from flask import Flask
```

```python
app = Flask(__name__)

@app.route("/")
def hello_world():
    return "<h1>Hello, World!!</h1>"

if __name__=="__main__":
    app.run(host="0.0.0.0")
```

<h3 style='text-align: left;'> 4.  Create a “/welcome” route to display the welcome message “Welcome to ABC Corporation” and a “/” route to show the following details: <br> Company Name: ABC Corporation <br> Location: India <br> Contact Detail: 999-999-9999 </h3>

```python
from flask import Flask

app = Flask(__name__)

@app.route("/Welcome")
def Welcome():
    return "<h1>Welcome to ABC Corporation</h1>"

@app.route("/Welcome/Details")
def Details():
    return "<p>Company Name : ABC Corporation <br> Location: India <br> Contact Detail: 999-999-9999</p>"

if __name__=="__main__":
    app.run(host="0.0.0.0")
```
